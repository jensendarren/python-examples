```python
from geopy.geocoders import Nominatim
import pandas
nom=Nominatim(user_agent="my-app-jupyter-test")
```

```python
# Try out the geocode API
nom.geocode("735 Dolores St, San Francisco, USA")
```

```python
# Read in our list of supermarket data
supermarkets=pandas.read_json('supermarkets.json')
```

```python
# Create a custom fields to store the full address to be used for geocoding
supermarkets['Full Address'] = supermarkets['Address'] + ", " + supermarkets['City'] + ", " + supermarkets['Country'] 
supermarkets
```

```python
# Now apply the call to nom.geocode too the new 'Full Address' field
supermarkets['geo']=supermarkets['Full Address'].apply(nom.geocode)
supermarkets
```

```python
# Extract out the lat and lon values from the geo object stored in the geo column
supermarkets['lat']=supermarkets['geo'].apply(lambda x: x.latitude if x!=None else None)
supermarkets['lon']=supermarkets['geo'].apply(lambda x: x.longitude if x!=None else None)
supermarkets
```
