```python
# Now we are going to practice slicing data
# First thing in this example is to set the index to 'Address' (since its unuque)
supermarkets=df1.set_index('Address')
supermarkets
```

```python
# Now we can slice the data based on the labels!
supermarkets.loc["735 Dolores St":"332 Hill St", "Country":"ID"]
```

```python
# Grab the cell data at a specific intercetion
supermarkets.loc["332 Hill St","Country"]
```

```python
# Get all rows (:) for a speccific column ('Country')
supermarkets.loc[:,"Country"]
```

```python
list(supermarkets.loc[:,"Country"])
```

```python
# Its more common to slice using indexes, like so
# Remember, slicing is 'upper bound exclusive'
supermarkets.iloc[1:4,1:4]
```

```python
# Get a single row of data
supermarkets.iloc[3, 1:4]
```

```python
# IX indexing is depreciated!
# supermarkets.ix[3,"Name"]
```

```python
# Drop a column by passing the index label of the column and axis index 1
supermarkets.drop("City", 1)
```

```python
# Drop a row by passing the index label of the row and axis index 0
supermarkets.drop("332 Hill St", 0)
```

```python
# This will drop the first 3 rows
supermarkets.drop(supermarkets.index[0:3],0)
```

```python
# This will drop the first 3 columns
supermarkets.drop(supermarkets.columns[0:3],1)
```

```python
# This works since index and columns are arrays on the dataframe
print(supermarkets.columns)
print(supermarkets.index)
```
